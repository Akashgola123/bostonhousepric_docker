```python
## Indenpendent and Dependent VAriable

x = dataset.iloc[:,:-2]

y = dataset.iloc[:,-1]
```

```python
## Train,test and split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
## standardize the dataset

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
x_train
```

MODEL TRANING


```python
from sklearn.linear_model import LinearRegression
```

```python
linear_reggresion = LinearRegression()
```

```python
linear_reggresion.fit(x_train,y_train)
```

```python
## print the cofficient and the intercept

print(linear_reggresion.coef_)
```

```python
print(linear_reggresion.intercept_)
```

```python
### on which parameter the model has been trained

linear_reggresion.get_params()
```

```python
## prediction with test data

reg_pred = linear_reggresion.predict(x_test)
```

```python
reg_pred
```

```python
## plot a scatter plot for the predicition

plt.scatter(y_test,reg_pred)
```

```python
residuals = y_test-reg_pred
```

```python
residuals
```

```python
## plots this residuals

sns.displot(residuals,kind='kde')
```

```python
## sclatter plot with respect to predicition and residuals

plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))

print(mean_squared_error(y_test,reg_pred))

print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

R Squared and ADJUTED R Square

```python
##R Square

from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
print(score)
```

```python
# R Adjusted Square 

1-(1-score)*(len(y_test)-1)/(len(y_test))- x_test.shape[1]-1
```
